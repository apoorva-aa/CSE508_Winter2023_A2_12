```python
# Pre-processing data

from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
from nltk.tokenize import WhitespaceTokenizer
from nltk.corpus import stopwords
import string
```

```python
def pre_process_data(main , flag = False):

    if (flag):
        print("-----------------------------------------------")
        print ("            Original Data")
        print("-----------------------------------------------")
        print("1 :" , main[0])
        print("2 :" ,main[1])
        print("3 :" ,main[2])
        print("4 :" ,main[3])
        print("5 :" ,main[4])


    # Step 1 - lowercasing

    lower_cased_data = []
    for sent in main:
        lower_cased_data.append(sent.lower())

    if(flag):
        print("-----------------------------------------------")
        print ("            After Lower Casing")
        print("-----------------------------------------------")
        print("1 :" ,lower_cased_data[0])
        print("2 :" ,lower_cased_data[1])
        print("3 :" ,lower_cased_data[2])
        print("4 :" ,lower_cased_data[3])
        print("5 :" ,lower_cased_data[4])

    # Step 2 - Tokenization

    token = WhitespaceTokenizer()

    tokenized_data = []
    for sent in lower_cased_data:
        tokenized_data.append(token.tokenize(sent))

    if(flag):
        print("-----------------------------------------------")
        print ("            After Tokenization")
        print("-----------------------------------------------")
        print("1 :" ,tokenized_data[0])
        print("2 :" ,tokenized_data[1])
        print("3 :" ,tokenized_data[2])
        print("4 :" ,tokenized_data[3])
        print("5 :" ,tokenized_data[4])

    
    # Step 3 - removing stop words

    stopWords = set(stopwords.words('english'))
    no_stop_data = []

    for li in tokenized_data:
        temp = []
        for word in li:
            if (word not in stopWords):
                temp.append(word)
        no_stop_data.append(temp)

    if(flag):
        print("-----------------------------------------------")
        print ("            After Stop Words removal")
        print("-----------------------------------------------")
        print("1 :" ,no_stop_data[0])
        print("2 :" ,no_stop_data[1])
        print("3 :" ,no_stop_data[2])
        print("4 :" ,no_stop_data[3])
        print("5 :" ,no_stop_data[4])

    # Step 4 - removing punctuations

    puncts = string.punctuation
    no_punc_data = []

    for li in no_stop_data:
        temp = []
        for word in li:
            temp_word = ""
            for alp in word:
                if (alp not in puncts):
                    temp_word += alp
            temp.append(temp_word)
        no_punc_data.append(temp)

    if(flag):
        print("-----------------------------------------------")
        print ("            After Punctuations removal")
        print("-----------------------------------------------")
        print("1 :" ,no_punc_data[0])
        print("2 :" ,no_punc_data[1])
        print("3 :" ,no_punc_data[2])
        print("4 :" ,no_punc_data[3])
        print("5 :" ,no_punc_data[4])

    # Step 5 - Removing white spaces

    final_data = []

    for li in no_punc_data:
        temp = []
        for word in li:
            if (word != ''):
                temp.append(word)
        final_data.append(temp)

    if(flag):
        print("-----------------------------------------------")
        print ("            After White Space removal")
        print("-----------------------------------------------")
        print("1 :" ,final_data[0])
        print("2 :" ,final_data[1])
        print("3 :" ,final_data[2])
        print("4 :" ,final_data[3])
        print("5 :" ,final_data[4])

    return final_data
```

```python
def unigram_index(data):
  index = {}
  for id in range(1400):
    tokens = data[id]
    for token in tokens:
      if token in index.keys():
        if(index[token][1][len(index[token][1])-1] == id):
           continue
        index[token].append(id)
        doc_freq = index[token][0]
        doc_list = index[token][1]
        doc_freq += 1
        doc_list.append(id)
        index[token] = [doc_freq,doc_list] 
      else:
        index[token] = [1,[1]]
  return index
```

```python
# Initializing the tfidf matrices

tfidf_bin = [[0.0 for _ in range(len(vocab))] for _ in range(len(final_data))]
tfidf_rc = [[0.0 for _ in range(len(vocab))] for _ in range(len(final_data))]
tfidf_tf= [[0.0 for _ in range(len(vocab))] for _ in range(len(final_data))]
tfidf_ln = [[0.0 for _ in range(len(vocab))] for _ in range(len(final_data))]
tfidf_dn = [[0.0 for _ in range(len(vocab))] for _ in range(len(final_data))]
```

```python
max_doc = {}

k = 0
for document in final_data:
  doc_words = set(document)
  max_f = 0
  for word in doc_words:
    max_f = max(max_f, document.count(word))
  max_doc[k] = max_f
  k += 1
```

```python
words = {}
for i, document in enumerate(final_data):
  for j, word in enumerate(vocab):
    words[word] = j
    # print(word)
    idf_val = idf[word]
    tf_bin = 0
    tf_rc = 0
    tf_tf = 0
    tf_ln = 0
    tf_dn = 0.5
    if word in document:
      tf_val = document.count(word)
      tf_bin = 1
      tf_rc = tf_val
      tf_tf = tf_val/len(document)
      tf_ln = math.log2(1+tf_val)
      tf_dn = (float((0.5+0.5*(tf_val)/max_doc[i])))

    tfidf_bin[i][j] = tf_bin*idf_val
    tfidf_rc[i][j] = tf_rc*idf_val
    tfidf_tf[i][j] = tf_tf*idf_val
    tfidf_ln[i][j] = tf_ln*idf_val
    tfidf_dn[i][j] = tf_dn*idf_val
```

```python
def print_docs(scores, scheme):
  scores.sort(reverse = True)
  res = scores[:5]

  print('Top 5 documents ranked by TF IDF ('+scheme+') are:')
  for doc in res:
    stri = "cranfield"
    temp = str(doc[1]+1)
    leng = len(temp)
    req = 4 - leng
    stri += "0" * req 
    stri += temp
    # stri += ".txt"
    # print(stri)
    print(stri,": ",doc[0])
  # print('\n')
```

```python
def get_tfidf_docs(query):
  query_vector = [0.0 for _ in range(len(vocab))]
  for p, word in enumerate(vocab):
    query_vector[p] = query.count(word)/len(query)
  
  doc_score_bin = []
  doc_score_rc = []
  doc_score_tf = []
  doc_score_ln = []
  doc_score_dn = []
  for i, document in enumerate(final_data):
    s_bin = 0
    s_rc = 0
    s_tf = 0
    s_ln = 0
    s_dn = 0
    for j, word in enumerate(vocab):
      s_bin += (tfidf_bin[i][j]*query_vector[j])
      s_rc += (tfidf_rc[i][j]*query_vector[j])
      s_tf += (tfidf_tf[i][j]*query_vector[j])
      s_ln += (tfidf_ln[i][j]*query_vector[j])
      s_dn += (tfidf_dn[i][j]*query_vector[j])
    doc_score_bin.append([s_bin,i])
    doc_score_rc.append([s_rc,i])
    doc_score_tf.append([s_tf,i])
    doc_score_ln.append([s_ln,i])
    doc_score_dn.append([s_dn,i])

  print_docs(doc_score_bin,'Binary')
  print_docs(doc_score_rc,'Raw Count')
  print_docs(doc_score_tf,'Term Frequency')
  print_docs(doc_score_ln,'Log Normalization')
  print_docs(doc_score_dn,'Double Normalization')
```

```python
def jaccard_coeff(query, document):
  intersection = query & document
  union = query | document
  return len(intersection)/len(union)
```

```python
def process_query(input):
  query = pre_process_data([input])
  # print(query)
  query = set(query[0])
  print(query)
  
  docs_best = []
  for doc in range(1400):
      coeff = jaccard_coeff(query, index[doc])
      docs_best.append([coeff, doc])
  
  docs_best.sort(reverse = True)
  return docs_best[:10]
```

```python
def input_query():
  query = input("Input query: ")
  res = process_query(query)
  print('Top 10 documents ranked by Jaccard coefficient are:')
  for doc in res:
    stri = "cranfield"
    temp = str(doc[1]+1)
    leng = len(temp)
    req = 4 - leng
    stri += "0" * req 
    stri += temp
    # stri += ".txt"
    # print(stri)
    print(stri,": ",doc[0])
```
